## Linear Classification


### Split dataset 
training--> 40

testing-->10


```python
#read data
IRIS = pd.read_csv('/content/Iris.csv' ) 

#insert b column 
IRIS.insert(0,"B",1)

#split according to species
IRIS_setosa=IRIS.loc[IRIS['species'] == 'Iris-setosa']
IRIS_versi=IRIS.loc[IRIS['species'] == 'Iris-versicolor']
IRIS_virgi=IRIS.loc[IRIS['species'] == 'Iris-virginica']

#split to train and test
IRIS_setosa_train,IRIS_setosa_test=train_test_split(IRIS_setosa, test_size=10)
IRIS_versi_train,IRIS_versi_test=train_test_split(IRIS_versi, test_size=10)
IRIS_virgi_train,IRIS_virgi_test=train_test_split(IRIS_virgi, test_size=10)
```

### Make Z for all feature


```python
#drop the species column to be easy to use  operations
y1=IRIS_setosa_train
y2=IRIS_versi_train
y3=IRIS_virgi_train
x1=y1.drop(columns='species')
x2=y2.drop(columns='species')
x3=y3.drop(columns='species')

#put all in one
Z1=np.concatenate((x1,x2,x3))
Z2=np.concatenate((x1,x2,x3))
Z3=np.concatenate((x1,x2,x3))

#create Z 
Z1[40:,:] = Z1[40:,:]*(-1)
Z2[:40,:] = Z2[:40,:]*(-1)
Z2[80:,:] = Z2[80:,:]*(-1)
Z3[:80,:] = Z3[:80,:]*(-1)
```

###Calculate w
## w=(Z^T * Z)^-1 * Z^T * b


```python
#W1:
r,c=np.shape(Z1)
q1=np.dot(Z1.transpose(),Z1)
s1=np.linalg.inv(q1)
b1 = np.ones((r, 1))
q2=np.dot(Z1.transpose(),b1)
W1=np.dot(s1,q2)

#W2:
r1,c1=np.shape(Z2)
q3=np.dot(Z2.transpose(),Z2)
s2=np.linalg.inv(q3)
b2=np.ones((r1, 1))
q4=np.dot(Z2.transpose(),b2)
W2=np.dot(s2,q4)

#W3:
r2,c2=np.shape(Z3)
q5=np.dot(Z3.transpose(),Z3)
s3=np.linalg.inv(q5)
b3=np.ones((r2,1))
q6=np.dot(Z3.transpose(),b3)
W3=np.dot(s3,q6)

```

### One Vs All 
Build 3 model


##w1, w2 , w3



```python
#W transpose:
W1_t = W1.transpose()
W2_t = W2.transpose()
W3_t = W3.transpose()

#create a new test matrix with droping the species column
i1=IRIS_setosa_test
i2=IRIS_versi_test
i3=IRIS_virgi_test
t1=i1.drop(columns='species')
t2=i2.drop(columns='species')
t3=i3.drop(columns='species')
P=np.concatenate((t1,t2,t3))

#if conditions for testing 
m,n=np.shape(P)
for i in range(m):
  if np.dot(W1_t, P[i,:])> 0 and np.dot(W3_t, P[i,:])< 0 and np.dot(W2_t, P[i,:])< 0:
   print("OHH Myyyy goooodddddd it is classs 1111")
  elif np.dot(W2_t, P[i,:])> 0 and np.dot(W1_t, P[i,:])< 0 and np.dot(W3_t, P[i,:])< 0:
   print("OHH Myyyy goooodddddd it is classs 22222222")
  elif np.dot(W3_t, P[i,:])> 0 and np.dot(W1_t, P[i,:])< 0 and np.dot(W2_t, P[i,:])< 0:
    print("OHH Myyyy goooodddddd it is classs 333333333")
  else :
    print("OHH Myyyy goooodddddd errorrrrrrrrr")
```
